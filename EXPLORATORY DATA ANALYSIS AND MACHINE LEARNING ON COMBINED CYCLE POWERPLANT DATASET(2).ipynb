```python
##MACHINE LEARNING MODEL (MULTIPLE LINEAR REGRESSION TO PREDICT THE ENERGY OUTPUT IN THE PLANT
#TARGET VARIABLE OR THE DEPENDENT VARIABLE= ENERGY OUTPUT
#FEATURE OR INDEPENDENT VARIABLES= Temp, Exha_Vaccum,Average pressure and relative humidity
```

```python
#Importing the libraries for the machine learning
```

```python
##DEFINING THE X and Y varibles
#We will drop the Energy output variable so that we remain with a dataframe of independent variables only and then get the values
X=DATASET.drop(['ENERGY OUTPUT'],axis=1).values
X
```

```python
#We will slice the out the energy output from the DATASET and get its values
Y=DATASET['ENERGY OUTPUT'].values
Y
```

```python
print(X)
```

```python
print(Y)
```

```python
##Splitting the data into training and test set(we will use the function from scikit library 
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=0)
```

```python
##Training the model on the training set
##We will import the linearr Regression model from scikil learn library
from sklearn.linear_model import LinearRegression
ml=LinearRegression()
ml.fit(X_train,Y_train)
```

```python
#PREDICTING THE TEST SET RESULTS
Y_PRED=ml.predict(X_test)
print(Y_PRED)
```

```python
ml.predict([[14.96,41.76,1024.07,73.17]])
```

```python
##EVALUATING THE MODEL
from sklearn.metrics import r2_score
r2_score(Y_test,Y_PRED)
```

```python
##VISUALIZING THE PREDICTED RESULTS
plt.figure(figsize=(15,10))
plt.scatter(Y_test,Y_PRED)
plt.xlabel('Y_TEST')
plt.ylabel('Y_PRED')
plt.title('ACTUAL VS PREDICTED')
```

```python
##PREDICTED VALUES
PREDICTED_VALUES_DATAFRAME=pd.DataFrame({'Actual_Values':Y_test,'Predicted_values':Y_PRED,'Difference':Y_test-Y_PRED})
PREDICTED_VALUES_DATAFRAME
```
