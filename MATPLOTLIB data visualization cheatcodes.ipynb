```python
##OBJECT ORIENTED APPROACH .
#We have seen the basic plotting however, it is recomended to use the object oriented aproach for more control and customization of ourplots.
##The idea behind the object oriented approach is that we create figure objects and then call in methods or attributes from that object.
##This elegant is very helpful when we are dealling with canvas that has multiple plots on it.
```

```python
##To start with in the first step, lets create a figure called fig 1
fig1=plt.figure()
```

```python
##STEP 2, WE ADD AXES to the figure object we created[left, bottom,width,height]
axes=fig1.add_axes([0.1,0.1,0.9,0.9])
```

```python
fig1
```

```python
#STEP3 PLOTTING THE DATA IN THE FIGURE
axes.plot(X,Y)
```

```python
##STEP 4, we add labels and titles as well
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Figure Title')
fig1
```

```python
##SUMMARY OF OBJECT ORIENTED METHOD
##1.Create an empty canvas . Fig1=plt.figure()
##2.ADD axes to the canvas    Fig1.add_axes([0.1,0.1,0.9,0.9])
##3.Plot in the canvas   .    Fig1.plot(X,Y)
##4.Set labels                Fig1.set_xlabel('')
##                            Fig1.set_ylabel('')
##                            Fig1.set_title('')
```

```python
##ADDING AN INSET PLOT USING OBJECT ORIENTED APPROACH
fig2=plt.figure()
axes_main=fig2.add_axes([0.1,0.1,0.9,0.9])
axes_inset=fig2.add_axes([0.3,0.5,0.3,0.4])
##MAIN
axes_main.plot(X,Y,'b*-')
axes_main.set_xlabel('main X label')
axes_main.set_ylabel('main Y label')
axes_main.set_title('main Title')

##INSET
axes_inset.plot(Y,X,'r--')
axes_inset.set_xlabel('Inset Xlebel')
axes_inset.set_ylabel('Inset Ylabel')
axes_inset.set_title('Inset title')
```

```python
#CREATING A FIGURE AND A SET OF SUBPLOTS.( Object Oriented Approach)
##We start by calling out the subplots
fig,axes=plt.subplots()
```

```python
#We then add the nrows and ncols
fig,axes=plt.subplots(nrows=2,ncols=2)
plt.tight_layout()
```

```python
##We can share x and yaxes within the subplots
fig,axes=plt.subplots(nrows=2,ncols=2,sharex=True,sharey=True)
plt.tight_layout()
```

```python
fig,axes=plt.subplots()
axes.plot(X,Y,'r--')
axes.set_xlabel('XLABEL')
axes.set_ylabel('YLABEL')
axes.set_title('Title')
```

```python
##HISTOGRAM
from random import sample
data=sample(range(1,1000),100)
plt.hist(data)
plt.xlabel('X axis title')
plt.ylabel('Y axis title')
plt.title('Histogram title')
```

```python
##Boxplot
data=[np.random.normal(0,std,100) for std in range (1,4)]
plt.boxplot(data, vert=True,patch_artist=True)
plt.xlabel('X axis title')
plt.ylabel('Y axis title')
plt.title('Boxplottitle')
```
