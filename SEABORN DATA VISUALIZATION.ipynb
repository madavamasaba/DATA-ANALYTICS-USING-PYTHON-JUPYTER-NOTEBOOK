```python
#Loading the tips dataset
tips=sns.load_dataset('tips')
tips
```

```python
#The data has a total bill along with the amount of tip that a customer left for the staff along.
#The most convenient way to take a quick look at a univariate (single variabe) distribution in seaborn is the distplot() function.
##The function will draw a histogram and fit a Kernel Distribution plot by default
##Note that a histplot() will also work
```

```python
##Drawing the histogram
sns.distplot(tips['total_bill'])
```

```python
##Drawing the histogram
sns.histplot(tips['total_bill'],kde=True,bins=30)
```

```python
##According to the plots, the data suggests that most bills are between 10 and 20 dollars
```

```python
##Joitplot()
##We want to do some comparison, how much the customer gave in tip based on their totalbill.
##seaborn provides a way of plotting a bivariate(two variable data)
##We use the jointplot() function which basically match up two distplots with our choice of what kind (Scatter by default) parameter to compare with.
##Other options include kind which = to either of these {"scatter"|"reg"|"resid"|"kde"|"hex"}
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
##The above jointplot is essentially just two distribution plots along with a scatter plot having tip on y and total bill on x axis
```

```python
##From the above plot,it can be seen that
#-Most of the tips are less than 4 dollars. The maximum tip is around 10 dollars
#-We can see some trend in the tips with total bill as well
#-We can see that the more bill a customer paid, the more tip he gave
```

```python
##Lets play with kind parameter. We can pass in the function a parameter kind='hex'. This will show a hexagon distribution representation.
#This is similar to scatter but useful ecspecially when you have several datapoints that are close together
```

```python
sns.jointplot(x='total_bill',y='tip',kind='hex',data=tips)
```

```python
##Lets use kind='reg'
#reg is for regression. With this, the plot is like a scatterplot with a regression line drawn by seaborn. 
#This is just showing almost like a linear fit to the scattered datapoints
##The plot has a pearson coefficient and a p-value.
```

```python
sns.jointplot(x='total_bill',y='tip',kind='reg',data=tips)
```

```python
##Kind=kde will give a 2D kde plot. Showing the darker region with density where most of the datapoint match up.
```

```python
sns.jointplot(x='total_bill',y='tip',kind='kde',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',kind='resid',data=tips)
```

```python
##PAIRPLOT
#Uses the function pairplot()
##We want to have a quick look on our entire dataset in order to see if there is a pairwise relationship
##This can be done using the seaborn function pairplot, which puts multiple pairwise bivariate distributions in a dataset together on a single canvas.
##This creates a matrix of axes and shows the relationship for eacch pair of columns on our dataset.
##By default, it also dwars a univariate distribution (histogram) of each variable on the diagonal axes.
```

```python
sns.pairplot(tips)
```

```python
##For larger dataframes, it takes longer for pairplots to draw.
##Along with plotting pairwise relationships across an aentire dataframe,( for numerical column), pairplot() supports color hue argument, for (categorical columns) as well. This is a great parameter.
##Lets display information for the categorical column sex for example
```

```python
sns.pairplot(data=tips, hue='sex')
```

```python
##CATEGORICAL PLOTS
##We are going to deal with categorical data in this lecture
```

```python
##Seaborn provides a wide range of options to plot categorical data
#1. Categorical scatterplots: they show each observation at each level of the categorical variable
##They include;-strippot()
#              ;-swarmplot()
#2.Categorical distribution plots: they show an abstract representation of each distribution of observations
##They include;-boxplot()
#             ;-violinplot()
#             ;boxenplot()
#3.Categorical estimateplot: Apply a statistical estimation to show a measure of central tendency and confidence interval
##They include;-barplot()
#             ;-countplot()
#             ;pointplot()
```

```python
#CATEGORICAL SCATTERPLOTS
##stripplot()
#stripplot() provides a simple way to show the values of some quantitative variables across a level of categorical variable.
#It draws a scatterplot where one of the variables is categorical
#Lets plot the total bill for each day in the week in our tips data to see how the stripplot will look like
```

```python
stripplot=sns.stripplot(x="day",y="total_bill",hue='sex',data=tips)
stripplot.legend(loc=0)
plt.title('STRIPPLOT SHOWING TOTAL BILLS PAID IN EACH DAY BY SEX')
```

```python
#Note that the hue parameter can also be applied in this plot.
#When using seaborn, a title is added using the function plt.title()
#A legend is added using the plotname.legend(loc=1,2,3,4 or 0)
#AS SHOWN ABOVE
```

```python
##SWARMPLOTS
#swarmplot() is similar to the stripplot() and uses different approach to avoid the overlapping issue in the datapoits by positioning each scatterplot point on the categorical acis with an algorithm that avoids overlapping points.
#This gives a better representation of the distribution of values.
#Lets replot the above but now using swarmplot()
```

```python
swarmplot=sns.swarmplot(x="day",y="total_bill",hue='sex',data=tips)
swarmplot.legend(loc=0)
plt.title('SWARMPLOT SHOWING TOTAL BILLS PAID IN EACH DAY BY SEX')
```

```python
swarmplot=sns.swarmplot(x="total_bill",y="day",hue='time',data=tips)
swarmplot.legend(loc=0)
plt.title('SWARMPLOT SHOWING TOTAL BILLS PAID IN EACH DAY BY time')
```

```python
##We can observe from the above plot that, 
#-Almost all customers or majority of customers come for lunch on thursdays
#-Majority come for dinner on weekends
#-On fridays, not many go for lunch nor dinner. They are maybe saving money for weekends spending.
```

```python
##CATEGORICAL DISTRIBUTION PLOTS
#1.Boxplot()
##Boxplots are used to show the distribution of categorical data.
##A boxplot is also known as a box and whiskerplot and shows the distribution of quantitative data in a way that facilitates comparison between variables or across levels of  categorical variable
##The boxes show the quartiles of the dataset whereas the whiskers show the rest of the distribution., except for points that are determined to be 'outliers',using a metod that is function of the inter-quartile range.
##In statistics, the quartiles of a ranked data values are three points that divide the data ino four equal groups, each comprising a quarter of the data.
##A quartile is a type of quartile. The first quartile(Q1) is defined as the middle number between the smallest number and the median number of the dataset.
#The second quartile(Q2) is the median of the data.
#The third quartile (Q3) is the middle value between the median and the highest value of the dataset.
```

```python
##Plotting boxplot
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
##Understanding the boxplot
#When we look at one of the box, undestand that;
#-Q1 is the distance between the bottom whisker to the bottom of the box.
#-Q2 is the distance between the bottom of the box to the line (median)
#-Q3 is the distance between the median to the end or upper of the box.
#-Q4 is the distance between the upper or end of the box to the top of the whisker
#-Points outside the box are outliers
```

```python
##We can introduce 'hue' to get more information on the box.
sns.boxplot(x='day',y='total_bill',hue='sex',data=tips)
```

```python
##Adding the hue parameter splits the boxes into two for each day based on their sex in different colors.
##Now we can compare daily behavior of the person along with other categorie sex or smoking.
```

```python
#CATEGORICAL ESTIMATE PLOTS
##Often, rather than showing the distribution within each category, you might want to show the central tendency of the values.
```

```python
#1.Barplots
##barplot() is a general plot that allows us to aggregate the categorical data based on some functions, by default , the mean.
##Note that the measures of central tendencies are often refered to as averages.
##The term central tendency dates fro the late 1920s.
##The most common measures of central tendencies are arithmetic mean, median and mode.
```

```python
#Lets try barplot() now
sns.barplot(x='day', y='total_bill',data=tips, hue='sex')
```

```python
##In the plot above, we get mean of the 'total_bill' paid by male in blue bar and female in orange bar.
##We can change the estimator to our own function, lets say we want to use the standard deviation as our estimator for gender on each day in the week.
##We can use numpy.std for this purpose
```

```python
sns.barplot(x='day', y='total_bill',data=tips,hue='sex', estimator=np.std)
```

```python
##COUNTPLOT
##This is univariate
#This is a special type of plot in statistics and it is used when we want o compare the number of observation in oe categorical variable that is in each category of the variable.
##This is prefered rather than computing a statistic for a second variable such as the mean,standard deviation and median.
##This is similar to a histogram over a categorical, rather than a quantitative variable.
##In seaborn, we do it using the countplot()
```

```python
sns.countplot(x="day",hue='sex',data=tips)
```

```python
##In the above category, the no:of observations in nested "sex" on each day of the week is along y
```

```python
sns.countplot(x='size',data=tips,hue='sex')
```
